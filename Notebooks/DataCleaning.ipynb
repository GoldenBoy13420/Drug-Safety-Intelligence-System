```python
def convert_to_years(row):
    code = row['age_cod'].strip().lower()

    if code == 'yr':
        return row['age']
    elif code == 'dec':
        return row['age'] * 10
    elif code == 'mon':
        return row['age'] / 12
    elif code == 'wk':
        return row['age'] / 52
    elif code == 'dy':
        return row['age'] / 365
    elif code == 'hr':
        return row['age'] / (365 * 24)
    else:
        return None
df['age_cod'] = df['age_cod'].fillna('').astype(str)
df['age'] = df.apply(convert_to_years, axis=1)
```
